```python
import pandas as pd
```

```python
rainfall = pd.read_csv("rainfall.csv")
flood = pd.read_csv("flood.csv")
urban = pd.read_csv("urban.csv")
```

```python
rainfall["is_synthetic_rainfall"] = False
rainfall["is_synthetic_rainfall"] = rainfall["is_synthetic_rainfall"].astype("bool")
```

```python
rainfall.head()
```

```python
flood.head()
```

```python
flood = flood.rename(
    columns={
        "year": "Year",
        "city": "City",
        "is_synthetic": "is_synthetic_flood",
    }
)
flood.head()
```

```python
urban.head()
```

```python
urban = urban.rename(columns={"Cities": "City", "is_synthetic": "is_synthetic_urban"})
urban.head()
```

```python
import pandas as pd

dataframes = [rainfall, flood, urban]
merged_df = dataframes[0]
for df in dataframes[1:]:
    merged_df = pd.merge(merged_df, df, on=["City", "Year"])
```

```python
merged_df.head()
```

```python
# Get the list of columns
cols = merged_df.columns.tolist()

# Find the indices of the columns to swap
idx1 = cols.index("Total Flood Extent (km^2)")
idx2 = cols.index("is_synthetic_flood")

# Swap the column names in the list
cols[idx1], cols[idx2] = cols[idx2], cols[idx1]

# Reindex the DataFrame with the new column order
merged_df = merged_df[cols]

merged_df.head()
```

```python
merged_df.to_csv("merged_data.csv", index=False)
```
