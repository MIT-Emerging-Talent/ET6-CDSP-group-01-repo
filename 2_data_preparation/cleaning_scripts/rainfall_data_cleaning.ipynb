```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/MIT-Emerging-Talent/ET6-CDSP-group-01-repo/refs/heads/main/1_datasets/raw_data/Rainfall_Data.csv"
```

```python
df = pd.read_csv(url)
```

```python
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
yearly_rainfall = df.groupby(["City", "Year"])["Rainfall_mm"].sum().reset_index()
yearly_rainfall
```

```python
yearly_rainfall.to_csv("yearly_rainfall.csv", index=False)
```
